```python
# Retrieving information from Twitter accounts and storing in pickle files

import tweepy
from twitter_keys import consumer_key, consumer_secret, access_token, access_secret
import pickle

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth)

PawanKalyan_tweets = api.user_timeline('PawanKalyan')

with open('PawanKalyan.pkl', 'wb') as f:
    pickle.dump(PawanKalyan_tweets, f)

VarunTej_tweets = api.user_timeline('IAmVarunTej')

with open('VarunTej.pkl', 'wb') as f:
    pickle.dump(VarunTej_tweets, f)

AlluArjun_tweets = api.user_timeline('AlluArjun')

with open('AlluArjun.pkl', 'wb') as f:
    pickle.dump(AlluArjun_tweets, f)
```
