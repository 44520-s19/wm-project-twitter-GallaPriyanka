## Twitter Accounts:

#### I am analyzing the following three twitter accounts:

1. PawanKalyan

```python
# Retrieving information from Twitter accounts and storing in pickle files

import tweepy
from twitter_keys import consumer_key, consumer_secret, access_token, access_secret
import pickle

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth)

# Pickling first twitter account - PawanKalyan
PawanKalyan_tweets = api.user_timeline('PawanKalyan')

with open('PawanKalyan.pkl', 'wb') as f:
    pickle.dump(PawanKalyan_tweets, f)

# Pickling second twitter account - IAmVarunTej

VarunTej_tweets = api.user_timeline('IAmVarunTej')

with open('VarunTej.pkl', 'wb') as f:
    pickle.dump(VarunTej_tweets, f)

# Pickling tecond twitter account - AlluArjun

AlluArjun_tweets = api.user_timeline('AlluArjun')

with open('AlluArjun.pkl', 'wb') as f:
    pickle.dump(AlluArjun_tweets, f)
```

```python
# Loading cached data from pickle files

import pickle

# Loading the first pickle file - PawanKalyan.pkl

with open('PawanKalyan.pkl', 'rb') as f:
    PawanKalyan_tweets = pickle.load(f)

print('Pawan Kalyans\'s Tweets:', PawanKalyan_tweets)

# Loading the second pickle file - VarunTej.pkl

with open('VarunTej.pkl', 'rb') as f:
    VarunTej_tweets = pickle.load(f)

print('Varun Tej\'s Tweets:', VarunTej_tweets)

# Loading the third pickle file - AlluArjun.pkl

with open('AlluArjun.pkl', 'rb') as f:
    AlluArjun_tweets = pickle.load(f)

print('Allu Arjun\'s Tweets:', AlluArjun_tweets)
```

# Description of analysis steps



```python
import pickle
from nltk_helpers import get_sentiments

# Sentiment Analysis for the first twitter account - PawanKalyan

print('The following is the sentiment analysis for Pawan Kalyan\'s tweets:')
for tweet in PawanKalyan_tweets:
    print(get_sentiments(tweet.text))

# Sentiment Analysis for the first twitter account - IAmVarunTej

print('\nThe following is the sentiment analysis for Varun Tej\'s tweets:')
for tweet in VarunTej_tweets:
    print(get_sentiments(tweet.text))
    
# Sentiment Analysis for the first twitter account - AlluArjun

print('\nThe following is the sentiment analysis for Allu Arjun\'s tweets:')
for tweet in AlluArjun_tweets:
    print(get_sentiments(tweet.text))
```

```python
# Presentation of my results using matplotlib

# Presentation of results for the first twitter account - PawanKalyan
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

sentiments = [get_sentiments(tweet.text) for tweet in PawanKalyan_tweets]

# build our xs, ys, and zs
xs = []
for sent in sentiments:
    xs.append(sent['neg'])

ys = [sent['neu'] for sent in sentiments]
zs = list(map(lambda x: x['pos'], sentiments))

ax.scatter(xs, ys, zs)

#add another set of data
import pickle
with open('arbys.pkl', 'rb') as f:
    arbys_tweets = pickle.load(f)

from nltk_helpers import split_sentiments

arbyX, arbyY, arbyZ = split_sentiments([get_sentiments(tweet.text) for tweet in arbys_tweets])
ax.scatter(arbyX, arbyY, arbyZ, color='r', marker='^')

ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_ylabel('Positive')

plt.show()
```
